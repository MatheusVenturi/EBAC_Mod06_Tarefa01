```python
estcivmae = {1:'Solteira', 2:'Casada', 3:'Viuva', 4:'Separado judicialmente/Divorciado', 5: 'União consensual (versões anteriores)', 9:'Ignorado'}
df_datasus['ESTCIVMAE'] = df_datasus['ESTCIVMAE'].map(estcivmae)

plt.bar(df_datasus['ESTCIVMAE'].value_counts().index, df_datasus['ESTCIVMAE'].value_counts())
plt.xticks(rotation=60)
```

03- APGAR5 médio por consultas

```python
df_datasus_ap_consultas = df_datasus[['APGAR5', 'CONSULTAS']]
df_datasus_ap_consultas[['CONS_04', 'CONS_03', 'CONS_02', 'CONS_01', 'CONS_09']] = 0

for i in df_datasus_ap_consultas.columns[1:2]:
  df_datasus_ap_consultas.loc[(df_datasus_ap_consultas[i].isin([4])), 'CONS_04'] = df_datasus_ap_consultas['APGAR5']
  df_datasus_ap_consultas.loc[(df_datasus_ap_consultas[i].isin([3])), 'CONS_03'] = df_datasus_ap_consultas['APGAR5']
  df_datasus_ap_consultas.loc[(df_datasus_ap_consultas[i].isin([2])), 'CONS_02'] = df_datasus_ap_consultas['APGAR5']
  df_datasus_ap_consultas.loc[(df_datasus_ap_consultas[i].isin([1])), 'CONS_01'] = df_datasus_ap_consultas['APGAR5']
  df_datasus_ap_consultas.loc[(df_datasus_ap_consultas[i].isin([9])), 'CONS_09'] = df_datasus_ap_consultas['APGAR5']

medias = []
for i in ['CONS_04', 'CONS_03', 'CONS_02', 'CONS_01', 'CONS_09']:
  medias.append(df_datasus_ap_consultas.loc[df_datasus_ap_consultas[i] != 0, i].mean())

plt.bar(['media_cons_04', 'media_cons_03', 'media_cons_02', 'media_cons_01', 'media_cons_09'], medias)
plt.xticks(rotation=30)
```

04- Painel de frequência - SEXO, GESTACAO, ESTCIVMAE, CONSULTAS

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
ax1.bar(df_datasus['SEXO'].value_counts().index, df_datasus['SEXO'].value_counts()/df_datasus['SEXO'].count())
ax1.set_title('Freq Sexo')
ax2.bar(df_datasus['GESTACAO'].value_counts().index, df_datasus['GESTACAO'].value_counts()/df_datasus['GESTACAO'].count())
ax2.set_title('Freq Gestação')
ax2.tick_params('x', labelrotation=30)
ax3.bar(df_datasus['ESTCIVMAE'].value_counts().index, df_datasus['ESTCIVMAE'].value_counts()/df_datasus['ESTCIVMAE'].count())
ax3.set_title('Freq Estado Civ Mãe')
ax3.tick_params('x', labelrotation=30)
ax4.bar(df_datasus['CONSULTAS'].value_counts().index, df_datasus['CONSULTAS'].value_counts()/df_datasus['CONSULTAS'].count())
ax4.set_title('Freq Consultas')
fig.tight_layout()
```
